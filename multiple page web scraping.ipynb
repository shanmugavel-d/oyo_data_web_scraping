**web scraping from oyo website and collecting data from the website**

```
# these are the empty lists where we store the data 

hotel_name = []               
hotel_id = []
hotel_cost = []
hotel_rating = []
hotel_address = []
hotel_amenity =[]
hotel_city = []
hotel_url=[]





#require important library to work on 
import requests                                # request used here for get url request in website
from bs4 import BeautifulSoup                  #bs4 is used for parse the htmal
import pandas as pd                            #pandas used for data frame and save as csv





#imporant u need to find your user agent on your system and this is used to avoid error 
head ={'User-Agent':"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36"}
url = "https://www.oyorooms.com/allcities/"    #this all citys url of oyo


#request getting response
response = requests.get(url , headers= head )  
response = response.content
soup = BeautifulSoup(response,"html.parser")    #response is parsed as html
c= soup.find_all("a", class_="c-kpzrw5")        #here we find a tag and class based on what information we need
city_name = []                                  #here we are save that information in empty list
city_links = []



for i in range(0,len(c)):
    city = c[i].text                            #here we are getting city name
    city_name.append(city)
    lis = c[i]["href"]                          #here we are getting final city link 
    links = "https://www.oyorooms.com/"+lis     #here we add orginal link and final city link
    city_links.append(links)

    
    

for h in  range(0,1):                       #here we are use range (0,349)  because now oyo provide city and len of the city link also 348
    city1 = city_name[h]                        #save city name
    url_du = city_links[h]                      #take particular link based on city
      
        
        
    #range 1,25 because the page starts with 1 and maxiam pages are 25    
    for i in range(1,25):                        
            url = f"{url_du}?page={i}"                       #to take information pagewise use for loop to add page number
            response = requests.get(url , headers= head )   #here get inforamation on html parse pagewise 
            response = response.content
            soup = BeautifulSoup(response,"html.parser")
            
            
            #here in this code we get name of the hotel 
            name = soup.find_all("h3")           
            for i in name:
                v1 = i.text
                hotel_name.append(v1)
                hotel_city.append(city1)
                #hotel_url.append(url_du)
              
            
             #in this code we get address of the hotel
            address = soup.find_all("div", class_="d-body-lg listingHotelDescription__hotelAddress")
            for i in address:                    
                v2 = i.text
                hotel_address.append(v2)
                
            
            #in  this code we get rating of the hotel
            rating = soup.find_all("div", class_="hotelRating") 
            for i in rating:
                v3 = i.text
                hotel_rating.append(v3)
                
            
            #in here we get cost of hotel room
            cost = soup.find_all("div", class_="listingPrice")    
            for i in cost:
                v4 = i.text 
                hotel_cost.append(v4)
               
                
            #in here we get id of the paticular hotel  
            ids = soup.find_all("a", attrs={"class":"c-nn640c u-width100"})   
            for i in ids:
                ids = i["href"]
                
                
                #in here we get website location   of the paticular hotel
                hotel_url.append(f"{url_du+ids}")  
                hotel_id.append(ids)
                
            
            #in here we get amenity of the hotel
            amenitys = soup.find_all("div", attrs={"class":"amenityWrapper"})
            for i in amenitys:                     
                amenity= i.text
                hotel_amenity.append(amenity) 
```

```
df = pd.DataFrame({"Hotel_name": hotel_name,
                   "Hotel_city":hotel_city,
                   "Hotel_url":hotel_url,
                   "Hotel_ids" : hotel_id,
                  "Hotel_address": hotel_address,
                  "Hotel_rating": hotel_rating,
                   "Hotel_amenitys": hotel_amenity
                  })
```
